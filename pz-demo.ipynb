```python
#read in data for analysis
import numpy as np
import csv
import re
baddists,gooddists = [],[]
badmags,goodmags = [],[]
badgmrs,goodgmrs = [],[]
goodspecs,goodphots = [],[]
with open('distinfo.csv', 'rb') as csvfile:
    distreader = csv.reader(csvfile, delimiter=' ')
    for row in distreader:
        baddists.append(map(float, re.findall('0.?[\d.]+',row[5])))#dists
        #line = row[5].split(' ')#dists
        #dist = [float(item) for item in line if (item!='' and item!=']' and item!='[' and item!='\n')]
        #baddists.append(dist)#dists
        badmags.append(float(row[1]))#rmags
        badgmrs.append(float(row[2]))#colors
with open('specinfo.csv', 'rb') as csvfile:
    distreader = csv.reader(csvfile, delimiter=' ')
    for row in distreader:
        gooddists.append(map(float, re.findall('0.?[\d.]+',row[6])))#dists
        goodspecs.append(float(row[4]))#speczs
        goodphots.append(float(row[5]))#photzs
        goodmags.append(float(row[1]))#rmags
        goodgmrs.append(float(row[2]))#colors
alldists = np.concatenate((baddists,gooddists))
allmags = np.concatenate((badmags,goodmags))
allgmrs = np.concatenate((badgmrs,goodgmrs))
allinfo = zip(alldists,allmags,allgmrs)
ngals,nspecs,ndists = len(allinfo),len(gooddists),len(baddists)
```

```python
#set up for probability distributions
import pyfits

#all p(z) share these bins
zbins = pyfits.open("http://data.sdss3.org/sas/dr8/groups/boss/photoObj/photoz-weight/zbins-12.fits")

#reformat bins
zlos = [zbins[1].data[i][0] for i in range(0,len(zbins[1].data))]
zhis = [zbins[1].data[i][1] for i in range(0,len(zbins[1].data))]
nbins = len(zbins[1].data)

#use centers of bins for plotting
zmids = [(zbins[1].data[i][0]+zbins[1].data[i][1])/2. for i in range(0,nbins)]

#use differences for plotting
zdifs = [zbins[1].data[i][1]-zbins[1].data[i][0] for i in range(0,nbins)]
```

```python
#prepare for distances
import math as m
import scipy as sp
from scipy import integrate

Om = 0.27
Ok = 0.
OL = 0.73

Ez = lambda z: m.sqrt(Om*(1+z)**3+Ok*(1+z)**2+OL)

DH = 3000./0.7*1e6#pc
arg = lambda z: DH/Ez(z)

DC = lambda z: integrate.quad(arg,0,z)[0]
```

```python
def lumdist(z):
    comoving = DC(z)
    d = comoving*(1+z)
    return d

def absmag(appmag,z):
    d = lumdist(z)
    distmod = 5*m.log10(d/10.)
    M = appmag-distmod
    return M

rmags = goodmags

specM = [absmag(rmags[i],speczs[i]) for i in range(0,nspecs)]
specMV = [absmag(rmags[i],speczs[i])/(4.*m.pi*lumdist(speczs[i])**3/3.) for i in range(0,nspecs)]
photM = [absmag(rmags[i],photzs[i]) for i in range(0,nspecs)]
photMV = [absmag(rmags[i],photzs[i])/(4.*m.pi*lumdist(photzs[i])**3/3.) for i in range(0,nspecs)]

#maxM = [absmag(rmags[i],maxz[i]) for i in range(0,ngals)]
expM = [absmag(rmags[i],expz[i]) for i in range(0,nspecs)]
expMV = [absmag(rmags[i],expz[i])/(4.*m.pi*lumdist(expz[i])**3/3.) for i in range(0,nspecs)]

distM = [[absmag(rmags[i],zmids[j]) for j in range(0,nbins)] for i in range(0,nspecs)]
distMV = [[absmag(rmags[i],zmids[j])/(4.*m.pi*lumdist(zmids[j])**3/3.) for j in range(0,nbins)] for i in range(0,nspecs)]
```

```python
Mbins = np.linspace(min(min(distM)),max(max(distM)),nbins+1)
Mmids = [(Mbins[k]+Mbins[k+1])/2. for k in range(0,nbins)]
allbins = []
allVbins = []
for k in range(0,nbins):
    eachbin = []
    eachVbin = []
    for i in range(0,nspecs):
        for j in range(0,nbins):
            if distM[i][j]>Mbins[k] and distM[i][j]<Mbins[k+1]:
                eachbin.append(pzdists[i][j])
                eachVbin.append(pzdists[i][j])
    allbins.append(eachbin)
    allVbins.append(eachVbin)
#pMdist = [sum([pzdist[i][j] if Mdist[i][j]>Mbins[k] and Mdist[i][j]<Mbins[k+1] for j in range(0,nbins)])for k in range(0,len(Mbins))] for i in range(0,ngals)]
Mfunc = [sum(allbins[j]) for j in range(0,nbins)]
MVfunc = [sum(allVbins[j])/(4.*m.pi*lumdist(zmids[j])**3/3.) for j in range(0,nbins)]
```

```python
photbins,specbins,expbins,probbins = [],[],[],[]
for k in range(0,nbins):
    photbin,specbin,expbin,probbin = 0,0,0,0
    for i in range(0,nspecs):
        if photM[i]>Mbins[k] and photM[i]<Mbins[k+1]:
            photbin=photbin+1/(4.*m.pi*lumdist(photzs[i])**3/3.)
        if specM[i]>Mbins[k] and specM[i]<Mbins[k+1]:
            specbin=specbin+1/(4.*m.pi*lumdist(speczs[i])**3/3.)
        if expM[i]>Mbins[k] and expM[i]<Mbins[k+1]:
            expbin=expbin+1/(4.*m.pi*lumdist(expz[i])**3/3.)
        if distM[i][j]>Mbins[k] and distM[i][j]<Mbins[k+1]:
            probbin=probbin+pzdists[i][j]/(4.*m.pi*lumdist(zmids[j])**3/3.)    
    photbins.append(photbin)
    specbins.append(specbin)
    expbins.append(expbin)
    probbins.append(probbin) 
```
