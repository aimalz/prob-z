```python
#set up for probability distributions
#all p(z) share these bins from Sheldon, et al.
zbins = pyfits.open("http://data.sdss3.org/sas/dr8/groups/boss/photoObj/photoz-weight/zbins-12.fits")

#reformat bins
zlos = [zbins[1].data[i][0] for i in range(0,len(zbins[1].data))]
zhis = [zbins[1].data[i][1] for i in range(0,len(zbins[1].data))]
nbins = len(zbins[1].data)

#use centers of bins for plotting
zmids = [(zbins[1].data[i][0]+zbins[1].data[i][1])/2. for i in range(0,nbins)]
zavg = sum(zmids)/nbins

#useful for plotting
zdifs = [zbins[1].data[i][1]-zbins[1].data[i][0] for i in range(0,nbins)]
zdif = sum(zdifs)/nbins
```

```python
seed_ngals = 1000
#define flat distribution for N(z)
avg_prob = 1./nbins/zdif
flat = seed_ngals*avg_prob
logflat = m.log(flat)
onetrueflatNz = np.array([flat]*nbins)
logonetrueflatNz = np.array([logflat]*nbins)
```

```python
ndim = 35
logmu = logonetrueflatNz[:ndim]
q=1.#0.5
e=0.1
tiny=1e-6
#covmat = [[q*m.exp(-0.5*e*(i-j)**2.) for j in range(0,ndim)] for i in range(0,ndim)]
covmat = np.array([[q*m.exp(-0.5*e*(i-j)**2.) for j in range(0,ndim)] for i in range(0,ndim)])+tiny*np.identity(ndim)
#priordist = sp.stats.multivariate_normal(mean=logmu,cov=covmat)
priordist = mvn(logmu,covmat)
```

```python
nwalkers = 80
iguess_dumb = priordist.sample_dumb(nwalkers)#[0]
iguess_legit = priordist.sample_legit(nwalkers)#[0]
iguess_test,rando_test = priordist.sample_test(nwalkers)#[0]
iguesses = [iguess_dumb,iguess_legit,iguess_test]
means = [logmu,logmu,rando_test]
ntests = len(iguesses)
setups = ['Prior Samples','Gaussian Ball Around Mean','Gaussian Ball Around Prior Sample']
```

```python
def sampling(n,t):
    sampler.reset()
    fracs,times = [],[]#,chains = [],[],[]#probs = []
    pos, prob, state = sampler.run_mcmc(iguesses[n], outiters[t])
    frac = sampler.acceptance_fraction
    fracs.append(np.median(frac))
    time = sampler.get_autocorr_time()
#    for at in range(0,ndim):
#        if np.isnan(time[at]):
#            time[at] = 100.
    times.append(np.median(time))
#    probs = sampler.lnprobability
#    for walk in sampler.chain:#nwalkers
#        chain = np.transpose(walk)
#        chains.append(chain)
    print(setups[n]+' '+str(outiters[t])+' complete')
    return fracs,times#,probs,chains
```

```python
def profiler(n,t):
    pr = cProfile.Profile()
    pr.enable()
    fracs,times = sampling(n,t)#,probs,chains
    pr.disable()
    s = StringIO.StringIO()
    sortby = 'time'
    ps = pstats.Stats(pr, stream=s).sort_stats(sortby)
    ps.print_stats()
    print s.getvalue()
    return fracs,times#,probs,chains
```

```python
def samplings(t):
    fracs,times = [],[]#,probs,chains = [],[],[],[]
    for n in range(0,ntests):
        frac,time = profiler(n,t)#,prob,chain
        fracs.append(frac)
        times.append(time)
        #probs.append(prob)
        #chains.append(chain)
    return fracs,times#,probs,chains
```

```python
def runner():
    fracs,times = [],[]#,probs,chains = [],[],[],[]
    for t in range(0,nruns):
        frac,time = samplings(t)#,prob,chain#outiters[t])
        fracs.append(frac)
        times.append(time)
        #probs.append(prob)
        #chains.append(chain)
        print(str(outiters[t])+' complete')
    return fracs,times#,probs,chains
```
